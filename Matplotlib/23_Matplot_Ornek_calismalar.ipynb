## Histogram

```python
plt.hist(x, bins = 50, color = "purple",edgecolor="cyan")
```

## Bar 

```python
x = np.array(["A","B","C","D"])
y = np.array([4,25,8,5])
```

```python
plt.bar(x,y)
```

```python
plt.bar(x,y, align = "edge", color = "green")
```

## Scatter

```python
x = np.array(np.random.randint(20,200,50))
```

```python
y = np.array(np.random.randint(50,300,50))
```

```python
sizes = np.array(np.random.randint(10,300,50))
```

```python
plt.scatter(x,y)
plt.grid()
plt.tight_layout()
```

```python
plt.scatter(x,y, c="red", edgecolor="black", s= sizes, alpha = .6)
```

```python
x = np.array(np.random.randint(20,200,50))
y = np.array(np.random.randint(50,300,50))
plt.scatter(x,y)

x = np.array(np.random.randint(20,200,50))
y = np.array(np.random.randint(50,300,50))
plt.scatter(x,y)

x = np.array(np.random.randint(20,200,50))
y = np.array(np.random.randint(50,300,50))
plt.scatter(x,y)
```

## Pasta (Pie)

```python
x = [15,60, 22, 46]
```

```python
labels = "Dolar", "Altın", "Sterlin", "Lira"
```

```python
colors = "green","yellow", "cyan", "orange"
```

```python
plt.pie(x, labels=labels)
```

```python
plt.pie(x, labels=labels, colors=colors, startangle= 180, shadow = True, radius = .7)
```

```python
explode = (0.4,0,0,0)
```

```python
plt.pie(x, labels=labels, explode = explode, shadow=True)
```
